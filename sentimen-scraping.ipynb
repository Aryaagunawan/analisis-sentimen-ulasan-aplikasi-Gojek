```python
from google_play_scraper import reviews_all, reviews, Sort
import pandas as pd
```

Melakukan scraping ulasan dari aplikasi aplikasi Gojek versi 4.0.0 sampai 4.9.3

Menyimpan hasil menjadi file csv.

```python
# Mengambil 220000 ulasan aplikasi gojek
scrapreview = reviews(
    'com.gojek.app',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count = 220000,
    filter_score_with=5
)
```

```python
# Menyimpan hasil scrap ke dalam file csv
data = pd.DataFrame(scrapreview[0])
data.to_csv('ulasan-apk-gojek.csv', index=False)
```
